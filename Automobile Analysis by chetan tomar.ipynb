## Automobile Data Analysis

```python
# Improting required libraries.

import pandas as pd
import numpy as np
```

```python
# Reading Data csv file.

data = pd.read_csv("Automobile_data.csv")
df = pd.DataFrame(data)
print(df.info())
df.head()
```

```python
# Filling Blank cells for accurate output.

df['price'] = df['price'].fillna(0)
```

```python
# Que 1. what is the average price and average mileage for convertible cars.

convertible = df[df['body-style'] == 'convertible']
avg_mileage = convertible['average-mileage'].mean()
avg_price = convertible['price'].mean()

avg_result = pd.DataFrame({'Body-Type':["convertible"], 'Average-Mileage':[avg_mileage], 'Average-Price':[avg_price]}).reset_index(drop=True)
avg_result
```

```python
# Que 2. Sort cars by company and price in decending order.

sort = df.sort_values(by=['price'],ascending=False,ignore_index=True)
sort_company = sort[['company','price']]
sort_company
```

```python
# Que 3. Save data for 'Toyota' separately.

toyota = df[df['company'] == 'toyota']
toyota
```

```python
# Que 4. which company is producing most expensive car?

most_exp_car = sort_company.head(1)
most_exp_car
```

```python
# Que 5. Name the cars on basis of maximum length, highest wheelbase, maximum horsepower.

max_length = df['length'].max()
max_len_car = df[df['length']==max_length]

max_hp = df['horsepower'].max()
max_hp_car = df[df['horsepower']==max_hp]

max_wheelbase = df['wheel-base'].max()
max_wheelbase_car = df[df['wheel-base']==max_wheelbase]

final = pd.DataFrame({
    'Cars with Maximum Length': max_len_car['company'],
    'Max Length': max_len_car['length'],
    'Cars with Highest Wheelbase': max_wheelbase_car['company'],
    'Max Wheelbase': max_wheelbase_car['wheel-base'],
    'Cars with Maximum Horsepower': max_hp_car['company'],
    'Max Horsepower': max_hp_car['horsepower']
}).fillna(0)

final
```

```python
# Que 6. company wise average price of cars and order them in decending order.

avg_price = pd.DataFrame(df.groupby('company')['price'].mean().astype(int).sort_values(ascending=False),columns=['price']).reset_index()
avg_price
```

```python
# Que 7. create a column named: "decision", If wheel-base is less than 100 and length is less than 175 then "don't buy" else "buy".

df['Decision'] =  ['don\'t buy' if (wheel < 100 or length < 175) else 'buy' for wheel, length in zip(df['wheel-base'], df['length'])]
df
```

```python
# Que 8. Select only cars whose decision is "buy"; find average price based on company of car.

buy_avg = df[df['Decision']=='buy']
avg_car = pd.DataFrame(buy_avg.groupby('company')['price'].mean().astype(int)).reset_index()
avg_car
```
